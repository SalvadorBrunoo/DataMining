```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt

# Dados de exemplo
y =  np.array(df['PrecoVenda'].to_list())
X = df.drop("Nome", axis =1, inplace=True)
X = df.drop("PrecoVenda", axis =1, inplace=True)
X = df.values

# Separar o conjunto de dados em treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.05, random_state=10)

# Criar um modelo de regressão linear múltipla
modelo = LinearRegression()

# Treinar o modelo com os dados de treino
modelo.fit(X_treino, y_treino)

# Fazer previsões nos dados de teste
previsoes_teste = modelo.predict(X_teste)

# Calcular o coeficiente de determinação (R²) nos dados de teste
r2_teste = r2_score(y_teste, previsoes_teste)

# Calcular o erro quadrático médio (MSE) nos dados de teste
mse_teste = mean_squared_error(y_teste, previsoes_teste)

# Exibir os coeficientes e intercepto do modelo
print("Coeficientes:", modelo.coef_)
print("Intercept:", modelo.intercept_)

# Visualizar os resultados nos dados de teste
print("R² nos dados de teste:", r2_teste)
print("Erro Quadrático Médio nos dados de teste:", mse_teste)
```
